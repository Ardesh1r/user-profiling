# User Segmentation

we need to carefully select features that best represent user preferences and behavior. This will help in identifying meaningful clusters or segments within the user base.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
```

```python
df.columns
```

```python
# Step 1: Feature Selection
features = [
    'Age', 'Gender', 'spotify_usage_period', 'spotify_listening_device',
    'spotify_subscription_plan', 'preferred_listening_content', 'fav_music_genre',
    'music_lis_frequency', 'music_Influencial_mood', 'pod_lis_frequency', 'pod_variety_satisfaction'
]
```

```python
# Subset the DataFrame with the selected features
df_selected = df[features].copy()

# Encoding categorical variables
label_encoder = LabelEncoder()
for column in df_selected.columns:
    if df_selected[column].dtype == 'object':
        df_selected.loc[:, column] = label_encoder.fit_transform(df_selected[column])
        
# Scaling the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_selected)
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Range of clusters to try
cluster_range = range(2, 11)

# Elbow Method
wcss = []
silhouette_scores = []

for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(scaled_data)
    
    # Within-cluster sum of squares
    wcss.append(kmeans.inertia_)
    
    # Silhouette Score
    silhouette_avg = silhouette_score(scaled_data, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)

# Plot the Elbow Method
plt.figure(figsize=(12, 6))
plt.plot(cluster_range, wcss, marker='o')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()

# Plot the Silhouette Scores
plt.figure(figsize=(12, 6))
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette Scores For Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
df['Cluster2'] = kmeans.fit_predict(scaled_data)

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster3'] = kmeans.fit_predict(scaled_data)
```

```python
df
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# Reduce the data to 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Use the correct DataFrame that has the cluster labels
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=df['Cluster3'], palette='viridis', s=100)
plt.title('PCA Visualisation of KMeans Clusters (3 Clusters)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# Reduce the data to 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Use the correct DataFrame that has the cluster labels
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=df['Cluster2'], palette='viridis', s=100)
plt.title('PCA Visualisation of KMeans Clusters (2 Clusters)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```
